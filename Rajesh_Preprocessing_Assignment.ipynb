# Check and handle outliers in at least 3 columns in the dataset

```python
plt.boxplot(data['Age']) 
plt.title('Age')
```

```python
plt.hist(data['Age']) 
plt.show()
```

```python
Q1=np.percentile(data['Age'],25,interpolation='midpoint')
Q2=np.percentile(data['Age'],50,interpolation='midpoint')
Q3=np.percentile(data['Age'],75,interpolation='midpoint')
print(Q1), print(Q2), print(Q3)
```

```python
IQR = Q3 - Q1
print("IQR =", IQR)

low_lim = Q1 - 1.5*IQR
up_lim = Q3 + 1.5*IQR
print("low_lim =", low_lim)
print("up_lim =", up_lim)
```

```python
outliers_age = []
for i in data['Age']:
    if (i>up_lim) or (i<low_lim):
        outliers_age.append(i)
print("Outliers in Age column are ",outliers_age)
print(len(outliers_age))
```

```python
ind_up_lim=data['Age']>up_lim
ind_low_lim=data['Age']<low_lim
outl_1=data.loc[ind_up_lim].index
outl_2=data.loc[ind_low_lim].index
```

```python
data_pre=data.drop(outl_1)
data_pre=data_pre.drop(outl_2)
print('Data post removal of age outliers')
data_pre
```

# Do min max scaling on the feature set (Take ‘Survived’ as target)

```python
from sklearn import preprocessing as pp
from sklearn.model_selection import train_test_split
```

```python
data.columns
```

```python
data.head()
```

```python
data1=data.drop(['Name', 'Sex','Ticket','Embarked'],axis=1)
data1
```

```python
# Generate dataset
X=data1.drop(('Survived'),axis=1)
y=data1['Survived']
```

```python
# split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y,random_state=42,test_size=0.30)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_reg=LinearRegression()
model=lin_reg.fit(X_train,y_train)
predictions=model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
print('The mean squared error value is',mean_squared_error(y_test,predictions))
print('The r squared value is',r2_score(y_test,predictions))
```
